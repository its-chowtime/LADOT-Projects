```python
import time
def sleeper():
    while True:
        num = input('Seconds: ')
        try:
            num = float(num)
        except ValueError:
            print('Enter in a number.\n')
            continue
            
        print('Before: %s' % time.ctime())
        time.sleep(num)
        print('After: %s\n' % time.ctime())
        
```

```python
try:
    sleeper()
except KeyboardInterrupt:
    print('\n\nKeyboard exception received. Exiting.')
    exit()
```
